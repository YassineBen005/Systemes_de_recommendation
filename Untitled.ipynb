```python
import numpy as np

# --- Paramètres Globaux ---
max_iterations = 10  # Réduit pour tester plus vite (remettre à 20 si besoin)
lambda_reg = 0.1

# --- Fonctions Utilitaires ---

def read(filename):
    return np.loadtxt(filename, dtype=int)

def erreur(matrice_reel, matrice_predit, liste_idx):
    """
    Calcule le taux de réussite sur les indices cachés.
    """
    k = 0
    # On parcourt uniquement les indices cachés
    for elem in liste_idx:
        i, j = elem
        # On arrondit la prédiction à l'entier le plus proche (ex: 1.6 -> 2)
        note_predite = int(round(matrice_predit[i, j]))
        note_reelle = int(matrice_reel[i, j])
        
        if note_predite == note_reelle:
            k += 1
            
    if len(liste_idx) > 0:
        perc = (k / len(liste_idx)) * 100
        print(f"Taux de réussite : {perc:.2f} %")
    else:
        print("Aucun indice à tester.")

def cost(U, V, R, lambda_reg):
    """Calcule la fonction de coût (Erreur quadratique + Régularisation)"""
    m, n = R.shape
    cost_val = 0
    # Attention: Ces boucles sont lentes en Python pur
    for i in range(m):
        for j in range(n):
            if R[i, j] != 9:
                cost_val += (R[i, j] - np.dot(U[i], V[j]))**2 
    
    cost_val += lambda_reg * (np.linalg.norm(U, 'fro')**2 + np.linalg.norm(V, 'fro')**2)
    return cost_val

# --- Fonctions de Mise à Jour (ALS) ---

def update_U(U, V, R, lambda_reg, k):
    m, n = R.shape
    # Pour chaque utilisateur i
    for i in range(m):
        # On construit le système linéaire A * U[i] = B
        A = np.zeros((k, k))
        B = np.zeros(k)
        
        for j in range(n):
            if R[i, j] != 9:  # Si l'item j a été noté par i
                A += np.outer(V[j], V[j])
                B += R[i, j] * V[j]
        
        # Ajout de la régularisation (Matrice identité * lambda)
        A += lambda_reg * np.eye(k)
        # Résolution du système
        try:
            U[i] = np.linalg.solve(A, B)
        except np.linalg.LinAlgError:
            pass # Si la matrice est singulière (rare avec régularisation)
            
    return U

def update_V(U, V, R, lambda_reg, k):
    m, n = R.shape
    # Pour chaque item j
    for j in range(n):
        A_prime = np.zeros((k, k))
        B_prime = np.zeros(k)
        
        for i in range(m):
            if R[i, j] != 9:  # Si l'utilisateur i a noté j
                A_prime += np.outer(U[i], U[i])
                B_prime += R[i, j] * U[i]
        
        A_prime += lambda_reg * np.eye(k)
        try:
            V[j] = np.linalg.solve(A_prime, B_prime)
        except np.linalg.LinAlgError:
            pass
            
    return V 

# --- Algorithme Principal ---

def ALS(R, epsilon):
    m, n = R.shape
    L_results = []
    
    # On teste avec k=10 facteurs latents
    for k in [10]:
        print(f"--- Démarrage ALS avec k={k} ---")
        # Initialisation aléatoire (petite échelle)
        U = np.random.randn(m, k) * 0.1
        V = np.random.randn(n, k) * 0.1
        
        previous_cost = cost(U, V, R, lambda_reg)
        
        for iter in range(max_iterations):
            print(f"Itération {iter+1}/{max_iterations}...")
            
            U = update_U(U, V, R, lambda_reg, k)
            V = update_V(U, V, R, lambda_reg, k)
            
            current_cost = cost(U, V, R, lambda_reg)
            print(f"   Coût : {current_cost:.4f}")
            
            if abs(previous_cost - current_cost) < epsilon:
                print(f"Convergence atteinte à l'itération {iter+1}")
                break
            
            previous_cost = current_cost
        
        # On stocke la matrice complète prédite
        L_results.append(np.dot(U, V.T))
        
    return L_results 


# --- Exécution ---

# 1. Chargement
print("Chargement de la matrice...")
try:
    M_real = np.loadtxt("Matrix.txt", dtype=int)
except IOError:
    # Création d'une matrice fictive si le fichier n'existe pas pour tester le code
    print("Fichier Matrix.txt non trouvé, utilisation d'une matrice aléatoire.")
    M_real = np.random.randint(0, 3, size=(50, 20)) # 0, 1 ou 2

# 2. Masquer 10% (Test Set)
observed = np.argwhere(M_real != 9)
n_observed = len(observed)
n_to_hide = int(0.10 * n_observed)

print(f"Nombre de notes observées : {n_observed}")
print(f"Nombre de notes cachées pour le test : {n_to_hide}")

perm = np.random.permutation(n_observed)
hide_idx = observed[perm[:n_to_hide]]

L = M_real.copy()
for (i, j) in hide_idx:
    L[i, j] = 9  # masque = 9

# 3. Lancer ALS
# Note: epsilon 0.1 est assez large, 0.001 est plus standard
Predictions_List = ALS(L, 0.1) 

# 4. Évaluation
print("\n--- Résultats ---")
for mat in Predictions_List:
    # IMPORTANT: On compare la prédiction (mat) avec la VRAIE matrice (M_real)
    # sur les indices cachés (hide_idx)
    erreur(M_real, mat, hide_idx)
```
